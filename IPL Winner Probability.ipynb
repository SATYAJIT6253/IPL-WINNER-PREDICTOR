```python
import numpy as np;
import pandas as pd;
```

```python
match_df = pd.read_csv('matches.csv')
delivery_df = pd.read_csv('deliveries.csv')
```

```python
total_score_df = delivery_df.groupby(['match_id','inning']).sum()['total_runs'].reset_index();
print(total_score_df);
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1];
print(total_score_df);
```

```python
match_df = match_df.merge(total_score_df[['match_id','total_runs']],left_on ='id',right_on='match_id');
match_df
```

```python
print(match_df['team1'].unique());
```

```python
teams = [
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Royal Challengers Bangalore',
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Kolkata Knight Riders',
    'Rajasthan Royals',
    'Delhi Capitals',
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals');
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals');

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad');
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad');
```

```python
match_df = match_df[match_df['team1'].isin(teams)];
match_df = match_df[match_df['team2'].isin(teams)];
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied'] == 0];
match_df
```

```python
match_df['match_id'].unique
```

```python
match_df = match_df[['match_id','city','winner','total_runs']];
```

```python
match_df
```

```python
delivery_df = match_df.merge(delivery_df,on='match_id');
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2];
delivery_df
```

```python
delivery_df['balls_left'] = 126 - (6 * delivery_df['over']+delivery_df['ball']);
```

```python
delivery_df
```

```python
delivery_df['curr_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum();
```

```python
delivery_df['run_left'] = delivery_df['total_runs_x'] - delivery_df['curr_score']+1;
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0");
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x : x if x == "0" else "1");
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int);
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum().values;
delivery_df['wickets_left'] = 10 - wickets;
```

```python
delivery_df
```

```python
delivery_df['curr'] = (delivery_df['curr_score']*6)/(120 - delivery_df['balls_left']);
delivery_df['reqr'] = (delivery_df['run_left']*6)/(delivery_df['balls_left']);
```

```python
delivery_df
```

```python
def result(row):
  return 1 if(row['batting_team'] == row['winner']) else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','run_left','balls_left','wickets_left','curr','reqr','result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
from sklearn.model_selection import train_test_split 
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] > 0];

final_df
```

```python
X = final_df.iloc[:,:-1]
Y = final_df.iloc[:,-1]
print(Y.shape,X.shape)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, random_state=1,test_size=0.25)
print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python


pipe.fit(X_train,Y_train);
```

```python
Y_predict = pipe.predict(X_test);
from sklearn.metrics import accuracy_score
acc_score = accuracy_score(Y_predict,Y_test)
print(acc_score)
```

```python
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','run_left','balls_left','wickets_left','total_runs_x','curr','reqr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['run_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets_left'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
```
